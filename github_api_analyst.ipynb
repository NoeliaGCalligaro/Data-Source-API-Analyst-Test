```python
def buscar_repositorios(query="python", per_page=5):
    url = f"{BASE_URL}/search/repositories?q={query}&per_page={per_page}"
    response = requests.get(url, headers=HEADERS)
    data = response.json()
    return data

repos = buscar_repositorios()
print(json.dumps(repos, indent=2))
```

```python
def obtener_contenido(owner, repo, path="README.md"):
    url = f"{BASE_URL}/repos/{owner}/{repo}/contents/{path}"
    response = requests.get(url, headers=HEADERS)
    return response.json()

contenido = obtener_contenido("pandas-dev", "pandas")
print(json.dumps(contenido, indent=2))
```

```python
def paginar(endpoint, max_pages=3):
    resultados = []
    page = 1
    while page <= max_pages:
        url = f"{endpoint}&page={page}"
        response = requests.get(url, headers=HEADERS)
        if response.status_code != 200:
            break
        data = response.json()
        if not data:
            break
        resultados.extend(data)
        page += 1
        time.sleep(1)  # prevenir lmites
    return resultados
```

```python
def prueba_request(url):
    try:
        response = requests.get(url, headers=HEADERS)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.HTTPError as err:
        print(f"Error HTTP: {err}")
    except Exception as e:
        print(f"Otro error: {e}")
```
